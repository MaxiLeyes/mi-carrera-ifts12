```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos el dataset directo de internet
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Vemos las primeras 5 filas para chequear que cargó bien
df.head()
```

```python
# Info general: cuántas filas, qué tipo de dato, cuántos nulos
print("--- Información General ---")
df.info()

# Estadísticas rápidas: edad promedio, precio promedio, etc.
print("\n--- Estadísticas Descriptivas ---")
df.describe()
```

```python
# Sacamos el promedio de la columna Survived y lo pasamos a %
porcentaje = df['Survived'].mean() * 100

print(f'De {len(df)} pasajeros analizados, se salvó el {porcentaje:.1f}%')
```

```python
# Agrupamos por Sexo y sacamos el % de supervivencia
tabla_sexo = df.groupby('Sex')['Survived'].mean() * 100
print("Supervivencia por Género:")
print(tabla_sexo)

# Gráfico de barras
plt.figure(figsize=(8, 5))
tabla_sexo.plot(kind='bar', color=['pink', 'skyblue'], title='Supervivencia por Sexo')
plt.ylabel('% Supervivencia')
plt.xticks(rotation=0)
plt.show()
```

```python
# Agrupamos por Pclass (Clase)
tabla_clase = df.groupby('Pclass')['Survived'].mean() * 100
print("Supervivencia por Clase:")
print(tabla_clase)

# Gráfico de barras
plt.figure(figsize=(8, 5))
tabla_clase.plot(kind='bar', color='lightgreen', title='Supervivencia por Clase')
plt.xlabel('Clase (1 = Alta, 2 = Media, 3 = Baja)')
plt.ylabel('% Supervivencia')
plt.xticks(rotation=0)
plt.show()
```

```python
# Conclusiones del Análisis del Titanic

1. **Tasa de Supervivencia:** Solo el 38.4% de los pasajeros sobrevivió.
2. **Impacto del Género:** Las mujeres tuvieron una tasa de supervivencia de ~74% frente a un ~19% en hombres.
3. **Impacto Socioeconómico:** La clase social fue determinante; mientras que en 1ra clase se salvó el 63%, en 3ra clase solo lo hizo el 24%.
```
